```python
directory = '/Users/emilyflynn/Desktop/25M_tests/15M_logs'
p = getMaxRadiusProfile(directory)

m2s = [0.005, 0.1, 0.5, 1, 5, 7.5]

plotTConv(p, True, '15M')

for i in range(len(m2s)):
    s = str(m2s[i]) + 'M'
    plotTInspiral(p, m2s[i], s)

plt.xlim(1e9, 1e14)

plt.xlabel('Radius [cm]')
plt.ylabel('Time [s]')

plt.legend()
```
